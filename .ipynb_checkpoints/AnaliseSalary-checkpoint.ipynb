## Dividindo o salário anual para analisar o mensal 

```python
df["Monthly_Salary"] = df["Salary"] / 12
df.head()
```

## Salário Mensal Médio por Faixa Etária

```python
df["Faixa_Idade"] = pd.cut(df["Age"], bins=[18,25,35,45,55,65], 
labels=["18-25","26-35","36-45","46-55","56-65"])

media_por_idade = df.groupby("Faixa_Idade", observed=False)["Monthly_Salary"].mean()

media_por_idade.plot(kind='bar', figsize=(6,5))
plt.title("Salário Mensal Médio por Faixa Etária")
plt.xlabel("Faixa de Idade")
plt.ylabel("Salário Mensal Médio")
plt.tight_layout()
plt.show()
```

## Normalizando os textos de nível de educação

```python
df["Education Level"] = df["Education Level"].str.lower().str.strip()

df["Education Level"] = df["Education Level"].replace({
    "high school": "high school",
    "bachelor's degree": "bachelor",
    "bachelors degree": "bachelor",
    "bachelor's": "bachelor",
    "master's degree": "master",
    "masters degree": "master",
    "master's": "master",
    "phd": "phd",
    "phD": "phd",
})
```

## Salário Mensal Médio por Nível de Educação

```python
media_educacao = df.groupby("Education Level")["Monthly_Salary"].mean().sort_values()

media_educacao.plot(kind='bar', figsize=(6,5))
plt.title("Salário Mensal Médio por Nível de Educação")
plt.xlabel("Nível de Educação")
plt.ylabel("Salário Mensal Médio")
plt.tight_layout()
plt.show()
```
